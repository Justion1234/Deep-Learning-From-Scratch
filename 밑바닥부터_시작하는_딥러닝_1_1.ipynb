# 브로드캐스트

브로드 캐스트란?

numpy가 연산중에 다른 모양의 배열을 처리하는 방법

스칼라 : 단일차원의 값

벡터 : 한가지 스칼라 데이터타입으로 구성된 데이터 집합

```python
'''
브로드캐스트의  예) 10이라는 스칼라값 2 X 2행렬로 확대
'''

A = np.array([[1, 2], [3,4]])
a = np.array(10)
print(A * a)
```

원소 접근

```python
#원소의 인덱스 0부터 시작
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0]) #0행
print(X[0][1]) #(0, 1) 위치의 원소
```

```python
for row in X #for문으로 열 형태로 출력:
  print(row)
```

```python
X = X.flatten() #X를 1차원 배열로 변환(평탄화)
#그래서 3차원 배열 형태였던 X가 1차원 배열 형태로 변환되었음
print(X)
print(X[np.array([0, 2, 4])]) #1차원 배열 형태로 바꾼 X에서 인덱스가 0, 2, 4인 원소 얻기
```

```python
print(X > 15) #X = [51, 55, 14, 19, 0, 4]
print(X[X > 15]) #원소가 15넘는 것들만 출력
```

# Matplotlib

matplotlib란? 

그래프를 그려주는 라이브러리입니다. 그래서 matplotlib을 사용하면 그래프 그리기와 데이터 시각화가 쉬워집니다.

단순한 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt #pyplot은 그래프를 생성해주는 함수들을 가지고 있습니다.

#데이터 준비
x = np.arange(0, 6, 0.1) #0에서 6까지 0.1 간격으로 생성
y = np.sin(x) # y에다가 x의 각 원소에 넘파이의 sin함함수를 적용하여 할당

#그래프 그려주는 부분

plt.plot(x, y) #그래프를 그림
plt.show() #그래프를 화면에 출력
```

pylot의 기능

```python
import numpy as np
import matplotlib.pyplot as plt

#데이터 준비
x = np.arange(0, 6, 0.1) #0에서 6까지 0.1간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") #cos 함수는 점선으로 그리기, 그리고 linestyle이 선 모양 설정하는건데 점선으로 설정한겨
plt.xlabel("x") #x축 이름
plt.ylabel("y") #y축 이름
plt.title('sin & cos') #제목
plt.legend() # 범례 표시하기
# 범례란? 그래프에 데이터의 종류를 표시하기 위한 텍스트

plt.show #그래프 그려주기
```
